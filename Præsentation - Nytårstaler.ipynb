```python
import os                      # os tillader os bl.a. at finde filplaceringer på computeren
import numpy as np             # Numpy leverer noget af matematikken, der ligger under Pandas 
import pandas as pd            # Pandas tillader os at importere, oprette og manipulere data frames
from pandas import DataFrame   # Nogle libraries har under-biblioteker. Underbiblioteker importeres med from-kommandoen
```

```python
#Defineret og åbnet dokumentet med taler
txt_0 = open("Dronningenstaler.txt", "r")
DrM50 = txt_0.read()
```

```python
def rens_ord(text_0):
    text_1 = text_0.replace("\n"," ")
    text_2 = text_1.replace("."," ")
    text_3 = text_2.replace(","," ")
    text_4 = text_3.replace(":"," ")
    text_5 = text_4.replace("*"," ")
    text_6 = text_5.replace("–"," ")
    text_7 = text_6.replace("'"," ")
    text_8 = text_7.replace("”"," ")
    text_ren = text_8.replace("-"," ")
    text_lav = text_ren.lower()
    text_token = text_lav.split()
    return text_token
```

```python
def rens_sæt(text_0):
    text_1 = text_0.replace("\n"," ")
    text_2 = text_1.replace(","," ")
    text_3 = text_2.replace(":"," ")
    text_4 = text_3.replace("*"," ")
    text_5 = text_4.replace("–"," ")
    text_6 = text_5.replace("'"," ")
    text_7 = text_6.replace("”"," ")
    text_ren = text_7.replace("-"," ")
    text_lav = text_ren.lower()
    text_token = text_lav.split(".")
    return text_token
```

```python
ÅrListe = DrM50.split("Dronningens nytårstale") #inddeler teksten i de enkelte taler
```

```python
rensÅrOrd = []                     #opretter tom liste
for år in ÅrListe:                 #for hvert års tale... 
    rensÅrOrd.append(rens_ord(år)) #renses hvert ord, og hver tale tilføjes (år) til liste
```

```python
rensÅrOrd
```

```python
txtLen =[]                 #opretter tom liste 
for år in rensÅrOrd:      
    txtLen.append(len(år)) #Vi tæller antal ord i talerne.
```

```python
txtLen #viser antal ord pr tale
```

```python
AntalSæt = []                        
for år in ÅrListe:
    AntalSæt.append(rens_sæt(år))   #tæller antal sætninger
```

```python
SætÅr=[] # Viser antal ord pr sætning
for år in AntalSæt:
    SætÅr.append(len(år))
```

```python
#vi har nu fundet tre typer data: årsrække (udgivelsesår), sætningslængde (antal ord pr sætning) og txtLen (antal ord pr sætning)
#nu vil vi definere key og value
df_DM50 = pd.DataFrame ({"Årstal": årsrække, "Antal ord": txtLen, "Antal sætninger": SætÅr})
```
